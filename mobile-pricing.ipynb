```python
X = data_train
# y = data_train['price_range']
y = X.pop('price_range')
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.8)
```

```python
from xgboost import XGBRegressor, XGBClassifier
from sklearn.metrics import mean_absolute_error, accuracy_score, mean_squared_error, f1_score
```

```python
xgboost_regressor = XGBRegressor(n_estimators=800, learning_rate=0.01)
xgboost_regressor.fit(X_train, y_train)
```

```python
predictions = xgboost_regressor.predict(X_valid)
print(f"XGBOOST Regressor: {mean_absolute_error(predictions, y_valid)}")
```

```python
xgboost_classifier = XGBClassifier(objective='multi:softmax', num_class=len(set(y_train)), n_estimators=800, 
                                           learning_rate=0.1, max_depth=5, random_state=1)
xgboost_classifier.fit(X_train, y_train)
```

```python
predictions = xgboost_classifier.predict(X_valid)
print(f"XGBOOST Classifier: {accuracy_score(predictions, y_valid)}")
```

```python
from sklearn.neighbors import KNeighborsClassifier
knc = KNeighborsClassifier(n_neighbors=10)

knc.fit(X_train, y_train)
predictions = knc.predict(X_valid)

print(f"KNeighborsClassifier: {accuracy_score(predictions, y_valid)}")
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_valid = scaler.transform(X_valid)
```

```python
#Custom Classifier Logistic Regression
from classifier.logistic_regression import LogisticRegression

custom_lr = LogisticRegression(learning_rate=0.1, n_iters=14000)
custom_lr.fit(X_train , y_train)
y_pred = custom_lr.predict(X_valid)

print(f'Accuracy Score: {accuracy_score(y_pred, y_valid)}')
```

```python
#Classifier Logistic Regression from sklearn
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(X_train, y_train)

y_pred = lr.predict(X_valid)

print(f'Accuracy Score: {accuracy_score(y_pred, y_valid)}')
```
